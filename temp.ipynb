```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('corr.csv', sep=';')
df['WO [x1000]'] = df['WO [x1000]'].str.replace(',', '.').astype(float)
df
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
# Calculate the correlation between the two columns
corr_value = df['WO [x1000]'].corr(df['NL Beer consumption [x1000 hectoliter]'])

# Create the figure and the left y-axis
fig, ax1 = plt.subplots()

# Plot the first data set (WO [x1000]) on the left y-axis
line1, = ax1.plot(df['Year'], df['WO [x1000]'], label='WO [x1000]', color='blue')
ax1.set_xlabel('Year')
ax1.set_ylabel('WO [x1000]')
ax1.tick_params(axis='y')

# Create a second y-axis that shares the same x-axis
ax2 = ax1.twinx()

# Plot the second data set (NL Beer consumption) on the right y-axis
line2, = ax2.plot(df['Year'], df['NL Beer consumption [x1000 hectoliter]'], color='orange', label='NL Beer consumption')
ax2.set_ylabel('NL Beer consumption [x1000 hectoliter]')
ax2.tick_params(axis='y')

# Add the correlation value as text on the plot
ax1.text(0.02, 0.82, f'Corr: {corr_value:.2f}', transform=ax1.transAxes, fontsize=12,
         verticalalignment='top')

# Combine legends from both axes
lines = [line1, line2]
labels = [line1.get_label(), line2.get_label()]
ax1.legend(lines, labels, loc='upper left')

plt.savefig('solution.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
